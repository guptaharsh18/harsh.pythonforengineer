```python
#Case study 3 : HARSH GUPTA , 202401100700075
df = pd.read_excel('/content/Case_Study_5_B.xlsx')
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Read the uploaded Excel file
df = pd.read_excel('/content/Case_Study_5_B.xlsx')

# Step 3: Calculate percentage assuming marks are out of 15
df['Percentage'] = (df['Marks'] / 15) * 100

# Step 4: Categorize students
above_75 = df[df['Percentage'] > 75].reset_index(drop=True)
between_60_75 = df[(df['Percentage'] >= 60) & (df['Percentage'] <= 75)].reset_index(drop=True)
below_60 = df[df['Percentage'] < 60].reset_index(drop=True)

# Step 5: Pretty-print the categories
print("\nðŸ“˜ Students with >75%:\n")
print(above_75)

print("\nðŸ“™ Students with 60%-75%:\n")
print(between_60_75)

print("\nðŸ“• Students with <60%:\n")
print(below_60)

# Step 6: Enhanced Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Percentage'], bins=range(0, 101, 5), color='skyblue', edgecolor='black')
plt.xlabel('Percentage (%)', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.title('Histogram of Student Percentages', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(range(0, 101, 5))
plt.tight_layout()
plt.show()

# Step 7: Enhanced Scatter Plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df.index, df['Percentage'], c=df['Percentage'], cmap='viridis', edgecolors='black', s=70)
plt.xlabel('Student Number', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.title('Scatter Plot of Student Percentages', fontsize=14, fontweight='bold')
plt.colorbar(scatter, label='Percentage (%)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```
